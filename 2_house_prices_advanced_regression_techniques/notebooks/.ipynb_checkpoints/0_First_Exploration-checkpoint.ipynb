# House Pricing Case

This advanced regression process has many features to play with. we will try to break it down.

reference : # https://www.kaggle.com/serigne/stacked-regressions-top-4-on-leaderboard

```python
%matplotlib inline

## Import the important Package
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.ensemble import IsolationForest
```

```python
RAW_PATH="../data/raw/"
INTERIM_PATH="../data/interim/"
MODEL_PATH="../model/"
PROCESS_PATH="../data/processed/"
```

## Get The Data

```python
dtrain=pd.read_csv(RAW_PATH+'train.csv',encoding="latin-1")
dtest=pd.read_csv(RAW_PATH+'test.csv',encoding="latin-1")

dtrain.head()
```

```python
dtrain.head()
```

```python
## Number of data
len(dtrain)
```

```python
y_target0="SalePrice"
```

## Columns

```python
other_col=["KitchenAbvGr","BsmtFullBath","GarageCars","HalfBath",
           "BedroomAbvGr","TotRmsAbvGrd","FullBath","MSSubClass","BsmtHalfBath","Fireplaces"]
```

```python
## Categoric and Numeric (Heuristic)

categoric=dtrain.dtypes[dtrain.dtypes=="object"].index.tolist()
numeric=[]
for c in dtrain.columns:
    if c not in categoric and c.lower()!='id' and c!=y_target0:
        if (len(dtrain[c].unique())<15 or "class" in c.lower() or "year" in c.lower() or "yr" in c.lower()) and "overall" not in c.lower() and c not in other_col:
            categoric.append(c)
        else:
            numeric.append(c)
```

```python
# cat_col=['MSSubClass','MSZoning','Street','Alley','LotShape',
#          'LandContour','Utilities','LotConfig','LandSlope','Neighborhood',
#          'Condition1','Condition2','BldgType','HouseStyle','RoofStyle','RoofMatl','Exterior1st',
#          'Exterior2nd','MasVnrType','Foundation','Heating','CentralAir','Electrical','Functional',
#          'GarageType','GarageYrBlt','PavedDrive','Fence','MiscFeature','SaleType','SaleCondition']

# 'YearBuilt', 'YearRemodAdd', 'MoSold','YrSold' <- special

# 'ExterQual','ExterCond','BsmtQual','BsmtCond','BsmtExposure','HeatingQC',
# 'KitchenQual','FireplaceQu','GarageFinish','GarageQual','GarageCond'<-to numerical

# 'BsmtFinType1','BsmtFinType2','PoolQC' <-to numerical and multiple to it size
```

## Target Exploration

```python
the_target=dtrain[y_target0]
log_target=np.log(the_target)
```

```python
the_target.plot(kind="hist",figsize=(11,7))
```

```python
log_target.plot(kind="hist",figsize=(11,7))
```

```python
dtrain["log_target"]=log_target
```

```python
y_target1="log_target"
```

```python
dtrain_used=dtrain[["Id","log_target","SalePrice"]]
dtest_used=dtest[["Id"]]
```

## Numeric Simple Exploration

```python
add_cols=[y_target0,y_target1,"Id"]
dtrain_num=dtrain[numeric+add_cols]
```

```python
len_data=len(dtrain_num)
```

```python
## Find Features with Missing Value
data_num_des=dtrain_num.describe()
data_num_des
```

```python
## Handling Missing Value
num_handling={}
numeric2=np.copy(numeric).tolist()
col_num_miss=data_num_des.transpose()["count"]
col_num_miss=col_num_miss[col_num_miss<len_data].index.tolist()
for c in col_num_miss+["BsmtFullBath","BsmtFinSF1","TotalBsmtSF","GarageArea","GarageCars"]:
    med=np.percentile(dtrain[c].dropna(),50)
    dtrain[c+"_2"]=dtrain[c].fillna(med)
    dtest[c+"_2"]=dtrain[c].fillna(med)
    numeric2.remove(c)
    numeric2.append(c+"_2")
    num_handling[c]=med
    
dtrain_num2=dtrain[numeric2+add_cols]
```

```python
## Correlation all
corr_sale=dtrain_num2.corr().SalePrice**2
corr_sale=corr_sale.sort_values()
```

```python
corr_sale.plot(kind="barh",figsize=(11,7))
```

```python
## Select Features with high correlation

col_num0=corr_sale[corr_sale>0.05].index.tolist()

for cc in add_cols:
    try : 
        col_num0.remove(cc) 
    except: 
        pass
```

```python
dtrain_num3=dtrain_num2[["Id"]+col_num0]
X_num=dtrain_num3[col_num0]
X_num_test=dtest[col_num0]
```

```python
X_num.corr()
```

```python
## Find Outlier

outl=IsolationForest(contamination=0.05)
outl.fit(X_num)
dtrain_num3["outlier"]=outl.predict(X_num)

X_num2=dtrain_num3[dtrain_num3.outlier!=-1][col_num0]
```

```python
dtrain_num3.head()
```

```python
col_num0
```

```python
## Standardization
Scaler0=StandardScaler()
Scaler0.fit(X_num2)
col_num1=[c+"_ss" for c in col_num0]

dtrain_num3[col_num1]=pd.DataFrame(Scaler0.transform(X_num))
dtest[col_num1]=pd.DataFrame(Scaler0.transform(X_num_test))
```

```python
dtrain_used2=pd.merge(dtrain_used,dtrain_num3[col_num1+["Id"]],how="left",on="Id")
dtest_used2=pd.merge(dtest_used,dtest[col_num1+["Id"]],how="left",on="Id")
```

```python
dtrain_used2.head()
```

## Categoric Simple Exploration

```python
categoric_year=[c for c in categoric if "year" in c.lower() or "yr" in c.lower()]
categoric_common=[c for c in categoric if c not in categoric_year]
```

```python
categoric_year
```

### Common

```python
dtrain_cat_common=dtrain[categoric_common+add_cols]
dtest_cat_common=dtest[categoric_common+["Id"]]
```

```python
len(categoric_common+add_cols)
```

```python
## Find Features with Missing Value
col_cat_miss=dtrain_cat_common.isnull().sum()
col_cat_miss_f=col_cat_miss[(col_cat_miss>0)&(col_cat_miss<1000)]
col_cat_miss_m=col_cat_miss[col_cat_miss>1000]
```

```python
col_cat_miss_f
```

```python
col_cat_miss_m
```

```python
## handling Missing Value
categoric_common2=np.copy(categoric_common).tolist()

for c in col_cat_miss_f.index.tolist()+['Exterior1st','MSZoning','Functional','Utilities',
         'SaleType','KitchenQual','Exterior2nd']:
    dtrain[c+"_nn"]=dtrain[c].fillna("NADATA")
    dtest[c+"_nn"]=dtest[c].fillna("NADATA")
    categoric_common2.remove(c)
    categoric_common2.append(c+"_nn")
    
for c in col_cat_miss_m.index.tolist()+["PoolArea"]:
    dtrain[c+"_nn"]=np.where(dtrain[c].isnull().astype(int)==1,0,1)
    dtest[c+"_nn"]=np.where(dtest[c].isnull().astype(int)==1,0,1)
    categoric_common2.remove(c)
    categoric_common2.append(c+"_nn")
    
dtrain_cat_common2=dtrain[categoric_common2+["Id"]]
```

```python
dtrain_cat_common2.head()
```

```python
## Categoric Only without MoSold
categoric_only=list(set(categoric_common2)-set([c+"_nn" for c in col_cat_miss_m.index]+["PoolArea_nn"])-set(["MoSold"]))
categoric_only
```

```python
## handling categoric
# cc=categoric_only[10]
# print(cc)
dtrain[categoric_only].describe()
```

```python
# test_cat_miss=dtest[categoric_only].isnull().sum()
# test_cat_miss=test_cat_miss[test_cat_miss>0]
# test_cat_miss.index.tolist()
```

```python
OE1=OneHotEncoder()
OE1.fit(pd.concat([dtrain[categoric_only],dtest[categoric_only]]))
```

```python
dtrain[OE1.get_feature_names()]=pd.DataFrame(OE1.transform(dtrain[categoric_only]).toarray())
dtest[OE1.get_feature_names()]=pd.DataFrame(OE1.transform(dtest[categoric_only]).toarray())
```

```python
dtrain.head()
```

```python
dtrain_cat=dtrain[["Id"]+[c+"_nn" for c in col_cat_miss_m.index]+["PoolArea_nn"]+OE1.get_feature_names().tolist()]
dtest_cat=dtest[["Id"]+[c+"_nn" for c in col_cat_miss_m.index]+["PoolArea_nn"]+OE1.get_feature_names().tolist()]
```

```python
dtrain_used3=pd.merge(dtrain_used2,dtrain_cat,how="left",on="Id")
dtest_used3=pd.merge(dtest_used2,dtest_cat,how="left",on="Id")
```

### Year

```python
categoric_year
```

```python
## Year Data is changed into range data (numerical)
dtrain["Built2Sold"]=1/(dtrain.YrSold-dtrain.YearBuilt+1)
dtrain["Rem2Sold"]=1/(dtrain.YrSold-dtrain.YearRemodAdd+1)
dtrain["Gar2Sold"]=1/(dtrain.YrSold-dtrain.GarageYrBlt+1)
dtrain["is_Gar2Sold"]=np.where(dtrain.GarageYrBlt.isnull()==1,0,1)
dtrain["Gar2Sold"]=dtrain["Gar2Sold"].fillna(-1)

dtest["Built2Sold"]=1/(dtest.YrSold-dtest.YearBuilt+1)
dtest["Rem2Sold"]=1/(dtest.YrSold-dtest.YearRemodAdd+1)
dtest["Gar2Sold"]=1/(dtest.YrSold-dtest.GarageYrBlt+1)
dtest["is_Gar2Sold"]=np.where(dtest.GarageYrBlt.isnull()==1,0,1)
dtest["Gar2Sold"]=dtest["Gar2Sold"].fillna(-1)

Scaler1=StandardScaler()
Scaler1.fit(dtrain[["YrSold"]])
dtrain["YrSold_ss"]=Scaler1.transform(dtrain[["YrSold"]])
dtest["YrSold_ss"]=Scaler1.transform(dtest[["YrSold"]])

# dtrain[categoric_year+["YrSold_ss","Built2Sold","Rem2Sold","Gar2Sold","is_Gar2Sold"]].head()
# dtrain[categoric_year+["YrSold_ss","Built2Sold","Rem2Sold","Gar2Sold","is_Gar2Sold"]].head()
```

```python
dtrain_used4=pd.merge(dtrain_used3,dtrain[["Id","YrSold_ss","Built2Sold","Rem2Sold","Gar2Sold","is_Gar2Sold"]],
                      how="left",on="Id")

dtest_used4=pd.merge(dtest_used3,dtest[["Id","YrSold_ss","Built2Sold","Rem2Sold","Gar2Sold","is_Gar2Sold"]],
                      how="left",on="Id")
```

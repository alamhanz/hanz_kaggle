```python
%matplotlib inline

## Import the important Package
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.ensemble import IsolationForest
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
import xgboost as xgb

import warnings
warnings.filterwarnings('ignore')
```

```python
RAW_PATH="../data/raw/"
INTERIM_PATH="../data/interim/"
MODEL_PATH="../model/"
PROCESS_PATH="../data/processed/"
```

## Get The Data

```python
dtrain=pd.read_csv(PROCESS_PATH+'train_used0.csv',encoding="latin-1")
dtest=pd.read_csv(PROCESS_PATH+'test_used0.csv',encoding="latin-1")
```

```python
col_X=set(dtrain.columns.tolist())
col_X=list(col_X-set(['Id', 'log_target', 'SalePrice']))
col_X.sort()

col_Y=['log_target']
```

```python
X_tr=dtrain[col_X]
Y_tr=dtrain[col_Y]

X_te=dtest[col_X]
```

## Modelling

```python
X_train, X_test, y_train, y_test = train_test_split(X_tr, Y_tr, test_size=0.2, random_state=42)
```

```python
xg=xgb.XGBRegressor()
```

```python
# RF0=RandomForestRegressor(n_estimators=100, criterion='mse',
#                           max_depth=50, min_samples_split=5, min_samples_leaf=9,random_state=100)

# RF0=RandomForestRegressor(random_state=100)


# parameters = {'n_estimators':[100,150], 'max_depth':[50, 80],
#               'min_samples_split':[3,5,10],"min_samples_leaf":[3,5,10]}

parameters = {'nthread':[4], #when use hyperthread, xgboost may become slower
              'learning_rate': [0.05,0.1,0.15,0.2], #so called `eta` value
              'max_depth': [10,15,30],
              'min_child_weight': [5,10,15],
              'n_estimators': [10,15,20,50,100], #number of trees, change it to 1000 for better results
              'seed': [1337]}

clf = GridSearchCV(xg, parameters, cv=5, verbose=3)

clf.fit(X_train,y_train)
```

```python
# clf.best_estimator_
```

```python
# RF0=RandomForestRegressor(n_estimators=150, criterion='mse',
#                           max_depth=50, min_samples_split=3, min_samples_leaf=3,random_state=100)

RF0=xgb.XGBRegressor(learning_rate=0.3, max_depth=20, min_child_weight=5, n_estimators=20, seed=1337)
RF0.fit(X_train,y_train)
```

```python
result_train=pd.DataFrame({"real_tr":y_train.log_target.tolist(),"pred_tr":RF0.predict(X_train)})
result_train["err"]=result_train["real_tr"]-result_train["pred_tr"]
result_train["err2"]=result_train["err"]**2
np.sqrt(result_train.err2.mean())
```

```python
result_test=pd.DataFrame({"real_tr":y_test.log_target.tolist(),"pred_tr":RF0.predict(X_test)})
result_test["err"]=result_test["real_tr"]-result_test["pred_tr"]
result_test["err2"]=result_test["err"]**2
np.sqrt(result_test.err2.mean())
```

## TEST

```python
uplo=pd.DataFrame({"Id":dtest.Id,"pred_tr":RF0.predict(X_te)})
uplo["SalePrice"]=np.exp(uplo.pred_tr)
```
